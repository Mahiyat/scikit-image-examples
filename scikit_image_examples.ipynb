```python
# Using simple NumPy operations for manipulating images
import numpy as np
from skimage import data
import matplotlib.pyplot as plt

cat = data.cat()
cat[:10] = 0
mask = cat < 87
cat[mask] = 255
inds_x = np.arange(len(cat))
inds_y = (4 * inds_x) % len(cat)
cat[inds_x, inds_y] = 0

l_x, l_y = cat.shape[0], cat.shape[1]
X, Y = np.ogrid[:l_x, :l_y]
outer_disk_mask = (X - l_x / 2) ** 2 + (Y - l_y / 2) ** 2 > (l_x / 2) ** 2
cat[outer_disk_mask] = 0

plt.figure(figsize=(4, 4))
plt.imshow(cat, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Converting rgb to grayscale
from skimage import color
cat = data.cat()
cat_gray = color.rgb2gray(cat)
plt.figure(figsize=(4, 4))
plt.imshow(cat_gray, cmap='gray')
plt.axis('off')
plt.show()
```
